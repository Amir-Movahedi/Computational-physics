**Solve the schrodinger equation in real space with numberical method**

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
N = 500
Xmin = -10
Xmax = 10

h = (Xmax-Xmin)/N
x = np.arange(Xmin, Xmax, h)

c0 = 2/(h**2)
c1 = -1/(h**2)
#kinetic energy
T = c0 * np.diag(np.ones(N-1)) \
+c1*np.diag(np.ones(N-2), 1) + c1* np.diag(np.ones(N-2), -1)
# potential
V_0 = np.diag(x[1:,]**2)

H_0 = T+V_0
val, vec = la.eigh(H_0)
print(val[0:5])
```

```python
for i in range(3):
    plt.plot(vec[:, i]**2, label =f'$\psi_{i}$' )
plt.legend()
plt.grid()
```

**Potentail  $V(x) = \frac{ax}{L}$**

```python
#potential v(x) = ax/L
a = 10 #ev
L = 5 #A
V_1 = np.diag(x[1:,]*(a/L))
H_1 = T+V_1
val, vec = la.eigh(H_1)
print(val[0:5])
```

```python
for i in range(3):
    plt.plot(vec[:, i]**2, label =f'$\psi_{i}$' )
plt.legend()
plt.grid()
```
